```python
import tfrc
import numpy as np
import matplotlib.pyplot as plt
from librosa.display import specshow
from librosa import power_to_db
```

```python
signal, sr = tfrc.load("audio/sample_tcc.wav")
```

```python
spec_1024 = tfrc.stft_spec(signal, win_length=1024, n_fft=4096, hop_length=512)
specshow(power_to_db(spec_1024, ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear', cmap='inferno')
plt.ylim(0, 1000)
```

```python
spec_2048 = tfrc.stft_spec(signal, win_length=2048, n_fft=4096, hop_length=512)
specshow(power_to_db(spec_2048, ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear', cmap='inferno')
plt.ylim(0, 1000)
```

```python
spec_4096 = tfrc.stft_spec(signal, win_length=4096, n_fft=4096, hop_length=512)
specshow(power_to_db(spec_4096, ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear', cmap='inferno')
plt.ylim(0, 1000)
```

```python
swgm_spec = tfrc.swgm_from_specs((spec_1024, spec_2048, spec_4096))
specshow(power_to_db(swgm_spec, ref=np.max), sr=sr, hop_length=512, x_axis='time', y_axis='linear', cmap='inferno')
plt.ylim(0, 1000)
```
